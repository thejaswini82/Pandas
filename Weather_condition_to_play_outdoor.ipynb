```python
#2.Generate a Likelihood table by finding the probabilities of given features.
```

```python
precipitation_likelihood = weather_data['Data.Precipitation'].value_counts(normalize=True)

# Print the likelihood table
print("Likelihood Table for Data.Precipitation:")
print(precipitation_likelihood)
```

```python
# Calculate probabilities for the 'Date.Month' feature
likelihood_table = weather_data['Date.Month'].value_counts(normalize=True).reset_index()
likelihood_table.columns = ['Month', 'Probability']

likelihood_table = likelihood_table.sort_values(by='Month')

# Display the likelihood table
print("Likelihood Table for Date.Month:")
print(likelihood_table)
```

```python
#Now, use the Bayes theorem to calculate the posterior probability.
```

```python
# Calculate prior probabilities (assuming equal prior for each month)
total_samples = len(weather_data)
prior_probabilities = weather_data['Date.Month'].value_counts(normalize=True).to_dict()
```

```python
# Calculate likelihoods for each month
likelihoods = {}
for month in weather_data['Date.Month'].unique():
    subset = weather_data[weather_data['Date.Month'] == month]
    likelihoods[month] = len(subset) / total_samples
```

```python
# Calculate posterior probabilities using Bayes' theorem
posterior_probabilities = {}
for month in likelihoods:
    posterior_probabilities[month] = (likelihoods[month] * prior_probabilities[month]) / sum(
        [likelihoods[m] * prior_probabilities[m] for m in likelihoods]
    )
```

```python
print("Posterior Probabilities:")
print(posterior_probabilities)
```
